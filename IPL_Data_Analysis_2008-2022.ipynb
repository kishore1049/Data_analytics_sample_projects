```python
#Add a new column to extract only year from date using DatetimeIndex()
ipl['Season'] =pd.DatetimeIndex(ipl['Date']).year
ipl.head()
```

```python
#calculate matches played each season using diction
Matches_per_season = ipl.groupby(['Season'])['ID'].count().reset_index().rename(columns={'ID':'Matches'})
Matches_per_season
```

```python
#bar graph to plot visualize matches played in each season
sns.countplot(ipl['Season'])
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Season',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Total Matches played in each season ', fontsize=10,fontweight="bold")
plt.show()
```

```python
#merging two table using left join
season_data = ipl[['ID','Season']].merge(ball,left_on='ID',right_on='ID',how='left').drop('ID',axis=1)
season_data.head()
```

```python
#Total runs in each season
Season = season_data.groupby(['Season'])['total_run'].sum().reset_index()
p=Season.set_index('Season')
ax = plt.axes()
ax.set(facecolor = "grey")
sns.lineplot(data=p, palette = "magma")
plt.title('Total Runs in each season',fontsize=12,fontweight="bold")
plt.show()
```

```python
#Matches by runs/wickets/tie
ipl['WonBy'].value_counts()
```

```python
#No of toss won by each team
toss = ipl['TossWinner'].value_counts()
ax = plt.axes()
ax.set(facecolor ="grey")
sns.set(rc={'figure.figsize':(15,10)},style ='darkgrid')
ax.set_title('No of toss won by each team',fontsize=15,fontweight="bold")
sns.barplot(y=toss.index,x=toss,orient='h',palette="icefire",saturation=1)
plt.xlabel('# of tosses won')
plt.ylabel('Teams')
plt.show()
```

```python
#Toss decision
ax=plt.axes()
ax.set(facecolor="grey")
sns.countplot(x='Season',hue='TossDecision',data=ipl,palette="magma",saturation=1)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('\n Season',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Toss decision across seasons',fontsize=12,fontweight="bold")
plt.show()
```
